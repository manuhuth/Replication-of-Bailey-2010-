## 1.1 Summary of the Paper
--- 

In Bailey's [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) main analysis she tries to quantify the importance of the birth control pill for marital childbearing in the 1960s. The first birth control pill, the Envoid, had been introduced in 1957 and was followed by a start of a new era in US demographics in the 1960s. Lower fertility rates, of around $50\%$, and smaller family sizes have occurred and never changed back until today. Since fertility determines population growth, the age distribution and family structures, it is a main driver of the labor fource's composition and thus influences economic growth as a whole.\
The relationship between a new reliable contraception method and a decline in fertility seems straightforward on the first glimpse. However, previous analysis emphasized that, since women's education has improved in the 1960s, the opportunity costs for childbearing increased as well, leading to a decreasing demand for children. Moreover, the introduction period of the Envoid followed the US baby boom, which was expected to be succeeded by a natural decline. A similar decrease in fertility patterns was observable after the baby booms in the 1920, when no birth control pill was available. Therefore, it is not clear whether the 1960s were a 'contraceptive revolution' or a revolution of the times. The challenge of author was to distinguish the effect of the pill from the mentioned demographic, social and economic changes (Bailey, [2010](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98)).\
Building on the theoretical model of Michael and Willis [(1976)](https://www.nber.org/chapters/c3960.pdf), the approach of Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) is to make use of the heterogeneous Comstock Sales Bans Laws for the use of contraception methods in the early 1960s. Some states allowed contraceptive methods, some prohibited the advertising and some prohibited the use completely. She argues that the heterogeneity of these laws, if they were restrictive in a sense that they were taken seriously by the population, can be used to determine if the new oral contraception method had an impact on the structural change in childbearing. This differences can be accessed via the Griswold v. Connecticut decision of 1965. The US Supreme Court decided that Connecticut's ban for contraceptive is illegal and the state was forced to cancel it. In the aftermath of this decision, almost every state canceled their laws against the sale of contraceptives. To give the reader of this notebook a short introduction to the Comstock laws and the Griswold decision, section 1.1. summarizes the information given in Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98).  

To estimate the causal impact of the new oral contraception, Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) first applies different *Difference-in-Differences* (DiD) probit models to asses the impact of oral contraception on birth rates and subsequently linear DiD panel models to explore the relationship between sales bans and period fertility. The DiD approach is motivated by defining a model that can measure the average difference in the difference between states that had sales bans in force and states that no sales bans in force. If certain assumptions hold, it allows the researcher to determine whether a certain estimated treatment effect can be interpreted causal. Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) finds that the decline in marital fertility in the post 1960 era was driven by the contraceptive pill. She estimates that the effect was around $40\%$ of the total change in marital fertility.

## 1.1 History of Comstock Laws and Griswold v. Connecticut
---

The Comstock Laws were introduced in 1873 to outlaw, inter alia, printed material for *obscenities*. This included any method used for contraception. In 1920 45 states had the Comstock Laws as active laws in force. However, the statues varied across states. In general there were four different versions

1. General Obscenity: States that are classified as having general obscenity laws banned the spread of obscene information, sales of obscene information and sales of immoral *articles or instruments*. However, tools for contraception were not specifically mentioned.
2. Advertising or information bans: States that are classified as having advertising or information bans banned only advertising of information but sales were allowed. 
3. Sales ban: States classified as having sales bans prohibited Sales of any articles. 
4. Physician or pharmacists exceptions: States classified as having physician or pharmacists exceptions allowed sales only through physicians and sometimes pharmacists.
 
In 1965 Connecticut's law from 1879, which uniquely prohibited the 'use' of contraception was cancelled by the US Supreme Court. In response, state legislatures across the US responded by repealing or amending their obscenity statutes. By 1971 almost every state had eliminated its bans on contraception sales to married individuals. Further information can be found in the [legal appendix](http://www-personal.umich.edu/~baileymj/Bailey_Griswold_Legal_Appendix.pdf) of Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98).

# 2. Theoretical Background
---

The theoretical framework follows the work of Michael and Willis [(1976)](https://www.nber.org/chapters/c3960.pdf) and their model that extends the standard neoclassical approach of modeling the demand and supply for children. The number of children is modeled as a random variable, which is affected by the use of contraceptive methods. By choosing their contraception strategy, couples choose their monthly probability of conception. In the model this is equal to choose the distribution of a random variable. Michael and Willis [(1976)](https://www.nber.org/chapters/c3960.pdf) assume that the distribution is characterized by its mean and its variance. However, Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) focuses only on the first moment. Since this is not from major concern in Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98)'s analysis, this should not be too restrictive. Denoting

- $\alpha_j:=$ fixed cost of contraception (including searching for a supplier and learning about a method)
- $\beta_j:=$  marginal cost of contraception (behavioral costs e.g. abstinence, discomfort, necessarity of purchasing refills)
- $\mu_N:=$ mean of the expected distribution of children born without contraception

Couples maximize utility by weighing the marginal costs of averting each birth against the marginal benefit of attaining each ex ante distribution of childbearing outcome. Hence, couples optimize by choosing a distribution of childbearing outcomes $\mu^*$. The price of a contraception strategy $j$ is defined by $\pi_j = \alpha_j + \beta_j(\mu_N - \mu)$. Hence, the total cost function for achieving any birth distribution $\mu$ is given by

\begin{align}
C(\mu) = \min_j\{\alpha_j + \beta_j(\mu_N - \mu)\}.
\tag{2.1}
\end{align}

The introduction of the pill changed the total cost function. The fixed costs of the pill were, due to high amount of information that needed to be gathered and the more difficult search for a supplier, higher, or at least as high as for other contraceptive methods. However, since there was no interruption or discomfort at the time of the intercourse, the marginal costs of the pill are assumed to be smaller. Hence, conditioning on a constant demand for children, the reduced marginal costs of preventing conception should reduce the number of born children (Bailey, [2010]((https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98))), since individuals with a preference for low fixed costs would stick to their old method, achieving the same outcome as without the pill, and individuals with a preference for low marginal costs would either change their method of contraception or start to use contraception.\
Moreover, since the pill was more reliable than other contraceptive methods, it reduced the uncertainty of childbearing. This channel is complicated to interpret. On the one hand, greater reliability decreases the chances of receiving more babies than intended but on the other hand greater reliability reduces the risk of overshooting and therefore increases the number of babies. Hence, it cannot clearly be determined which effect rules out the other. Furthermore, the choice of contraception depends on the number of children that should be averted. Hence, the chosen method depends on the duration of use determined by the age of an individual. Due to the high fixed but low marginal costs, younger women should thus be more likely to use the pill after its introduction.

Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) extended the model to account for restrictive sales bans. She argues that the threat of penalties reduces the availability of all types of contraception. However, since the pill must be prescribed by pharmacists and physicians, its availability reduces even more because the latter have to lose more by conducting an illegal act. Moreover, unlike other contraceptive methods, the production of the pill relies on chemical components, which is likely to limit the black market production, at least in the short run. Additionally, one does not know whether the unknown chemicals of the black market pills are harmful or not, leading to even higher fixed and marginal costs in states with sales bans. To include sales bans, the cost function must thus be modified such that sales bans are allowed to have different influences on the cost function regarding the respective contraceptive method. This is done by multiplying both additively linked terms in the equation by a contraceptive varying rate greater or equal than one. The new price of a contraception strategy $j$ with sales bans in state $S$ is given by

$$
\pi^S_j = \alpha_j(1+t_{\alpha j}) + \beta_j(1+t_{\beta j})(\mu_N - \mu), 
\tag{2.2}
$$ 

whereby $  t_{\alpha_j}, t_{\beta_j} \geq 0 \forall j$  and for abstinence, withdrawal and rhythm: $t_{\alpha_j}, t_{\beta_j} = 0$. Following the stated arguments, $ t_{\alpha_{pill}}$ and $t_{\beta_{pill}}$ must be greater as for the other methods.
Unfortunately, birth control is even more complicated. Since the desired number of children reflects the demand (market prices, household income and preferences) and determines a woman's likelihood of adopting the pill, comparison of childbearing outcomes among women using and not using the birth control pill reveal little about the importance of demand or supply factors.

# 3. Extension | Econometric Background
---

To facilitate explanations of further analysis, I give a short review of the probit model. Since Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) uses a DiD probit model, which differs from the interpretation of the coefficients from an ordinary probit model, I also introduce how the marginal effects are computed and how they can be interpreted in the context of a DiD probit model.

## 3.1 Extension | Probit Model 
---

The probit approach models the conditional probability of a binary outcome variable $y_{it}$, conditioned on regressors $x'_{it}$ with indices $i$, representing the individual, and $t$, the time period when the observation was observed. Since Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) does not allow for temporal state-mobility of individuals, I have omitted a state index for ease of notation. Furthermore, it is assumed that there is a continuous latent process $y_{it}^*$ such that 


\begin{align}
y_{it} = \begin{cases}
1 & \text{for } y_{it}^* > 0 \\
0 & \text{for } y_{it}^* \leq 0\\
\end{cases}.
\tag{3.1}
\end{align}


Furthermore it is assumed that the conditional mean of the latent process is a linear function of parameters $\beta \in \mathbb{M}_{p \times 1}$ and that there are conditional independent error terms $\epsilon_{it}|x \sim \mathcal{N}(0, \sigma^2_{it})$. Since the error term is dependent on each individual, a model robust against heteroscedasticity is feasible. Armed with these assumptions, we can derive the conditional probability of $y_{it}$ as being

\begin{align}
P(y_{it} =1|x) = P(y_{it}^* > 0| x) = P(x'_{it} \beta + \epsilon_{it} > 0| x) = P(\epsilon_{it} > - x'_{it} \beta| x) = 1 - \Phi_{it}(-x'_{it} \beta) = \Phi_{it}(x'_{it} \beta), \tag{3.2}
\end{align}

whereby $\Phi_{it}$ is the cumulative normal distribution with mean zero and variance $\sigma^2_{it}$. The transformation of the last equality sign comes from the symmetric properties of the normal distribution with mean zero.

Since Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) uses regional clustering, it might be that $\text{E}(\epsilon_{it} \cdot \epsilon_{jt}|x) \neq 0$, if $i$ and $j$ are from the same region. Thus, the errors are not conditionally independent and traditional maximum likelihood is strictly spoken infeasible. However, Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) used the *probit* command and the *cluster()* option in Stata to compute her results. It can be found on the [Stata help desk](https://www.stata.com/support/faqs/statistics/maximum-likelihood-estimation/) that this command treats the data as if it was independent by expressing the estimator $\hat{\beta}$ as the argument minimum of the negative (*incorrect*) log-likelihood function. 

\begin{align}
\hat{\beta} =  \arg\min_{b \in \mathbb{M}_{p \times 1}} [- \sum_{(i,t) \in I}  (y_{it} \ln[\Phi_{it}(x'_{it} b)] + (1-y_{it}) \ln[1-\Phi_{it}(x'_{it} b)])]. \tag{3.3}
\end{align}

Whereby $I$ is an index set containing all tuples of individuals and time points. Note that usually the solution cannot be obtained in closed form and therefore numerical methods must be applied. Since Stata was used to construct the results from the paper but I used Python, it is possible that the results differ slightly at some digits. Nevertheless, the results should qualitatively stay the same. I check the results for different starting values in section 7. Unfortunately, I was not able to find out which optimizer Stata uses by default for the probit model. Hence, the extension section contains validity checks for different optimizers yielding a broad range of results.

## 3.2 Extension | Average Marginal Effects (AME)
---

Using the above formulas, a specific parameter $\beta_j$ from the vector $\beta$ can be interpreted as the effect from a marginal increase of the $j$-th variable on the respective z-score, since the z-score is a linear function in $\beta$. However, the interpretation of $\beta_j$ regarding the probability $P(y_{it} =1|x)$ is not similarily straightforward, since it depends on the magnitude of the z-score and therefore also on the values of other covariates. Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) uses the user written *margeff* (Bartus, [2005](https://journals.sagepub.com/doi/pdf/10.1177/1536867X0500500303)) command from Stata. This command estimates *average marginal effects* (AME), that is to say, the changes in probability of $y_{it}$, when $x_{it,j}$ is increased by one. The reported marginal effects are sample averages of this changes. Hence, the AME for variable $j$ is given by

$$
\begin{align}
AME_j = \frac{1}{N} \sum_{(i,t) \in I} [ P(y_{it} = 1|x_{it,(-j)}, x_{it,j} = x'_{it,j} + 1) - P(y_i = 1|x_{it,(-j)}, x_{it,j} = x'_{it,j}) ], \tag{3.4}
\end{align}
$$

whereby $I$ is again an index set containing all tuples of individuals and time points. $N$ is the total number of observations. Note that $x'_{it,j} = 0$, if the $j$-th regressor is a dummy. The command *get_margeff* from Statsmodels applies the same algorithm to compute the marginal effects and therefore the results I obtained for the point estimates of the marginal effects and the results from Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) are comparable.\
The standard errors for AME can be approximated by approximating the variance-covariance matrix with $G V G'$, whereby $G$ is a matrix containing the derivatives of marginal effects with respect to the parameter estimates and $V$ is the covariance matrix of parameter estimates (Bartus, [2005](https://journals.sagepub.com/doi/pdf/10.1177/1536867X0500500303)). However, I did not find a source explicitly stating how the marginal standard errors are computed, which could explain small differences in marginal standard errors of the paper and the ones I obtained. Neither for Statsmodels nor for Stata. Since this is not of major concern for my replication, I will not go further into detail here. 

## 3.3 Extension | DID Probit Model
---

To keep the Identification section short, I will introduce the methodology of DiD (Probit) models, which is used in section 4 to derive causal identification. Note that in the general derivation of the DiD method, two groups are untreated in the first period and one of them is treated in the second period. However, with one treated group in the first period and no treated group in the second period, the application of Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) is vice versa. Hence, I will derive the the DiD estimator for the *reverse* case. There is a recent paper by Kim and Lee [(2019)](https://www.researchgate.net/publication/325570354_Difference_in_differences_in_reverse) that fully discusses the issues of the *reverse* DiD. 

Individuals are either part of the treatment group $g_T$ or of the control group $g_C$. There are two time periods $t_1$ and $t_2$. Including dummies indicating the respective group, time and an interaction term of these dummies, $y^*_{it}$ can be written as

$$ y_{it}^* = \alpha_0 + \gamma_{tr} \cdot D_{g_T} + \gamma_{ti} \cdot D_{t_2} + \gamma_{int} \cdot D_{g_T}D_{t_2} + x_{it}\beta + \epsilon_{it}, \tag{3.5}$$ 

whereby $D_{g_T}$ is one, if individual $i$ is from the treatment group, $ D_{t_2}$ is one, if the observation is from the second period, $\alpha_0$ is the intercept and $x_{it}$ denotes additional covariates. Applying the difference of the conditional expectations between the groups in the first period yields the expected difference in the z-score between the treatment and the non-treatment group in the first period 

$$ E(y_{it}^*| D_{g_T} = 1, D_{t_2} = 0, x_{it}) - E(y_{it}^*| D_{g_T} = 0, D_{t_2} = 0, x_{it}) = \alpha _0 + \gamma_{tr} + x_{it}\beta - (\alpha _0 + x_{it}\beta) = \gamma_{tr}. \tag{3.6}$$ 

The same algebra is applied to find the expected difference in the z-score between the treatment group and the non-treated group in the second period. 

$$ E(y_{it}^*| D_{g_T} = 1, D_{t_2} = 1, x_{it}) - E(y_{it}^*| D_{g_T} = 0, D_{t_2} = 1, x_{it}) = \alpha _0 + \gamma_{tr} + \gamma_{ti} + \gamma_{int} + x_{it}\beta - (\alpha _0 + \gamma_{ti} + x_{it}\beta) = \gamma_{tr} + \gamma_{int}. \tag{3.7}$$

In the ordinary DiD it would now be necessary to subtract equation 3.6 from Equation 3.7. Since we are interested in the *reverse* effect, it is necessary to subtract equation 3.7 from equation 3.6 This results in the interaction term $\gamma_{int}$ multiplied by minus one. The latter is therefore the difference in the difference between the two groups in the respective periods in the z-score (Probit) or in $y_{it} = y^*_{it}$ (Panel). 

$$
(E(y_{it}^*| D_{g_T} = 1, D_{t_2} = 0, x_{it}) - E(y_{it}^*| D_{g_T} = 0, D_{t_2} = 0, x_{it})) - (E(y_{it}^*| D_{g_T} = 1, D_{t_2} = 1, x_{it}) - E(y_{it}^*| D_{g_T} = 0, D_{t_2} = 1, x_{it}))  = - \gamma_{int} \tag{3.8}
$$

When this effect can be interpreted as being causal is discussed in the identification section. The AME for the probit model and the panel equivalent are then computed analogously as in the standard models as described above. The extension to models with more than one 'DiD-term' are straightforward by including the additional terms in $X_{it}$.

# 4. Identification
--- 
The identification in Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) is done in three steps.

1. Did Comstock Sales Bans slowed the diffusion of oral contraception?
2. Was a possible diffusion driven by a change in the demand of children caused by the sales bans?
3. The relationship between sales bans and period fertility

The first two steps establish the relevance and the validity of using the statuary bans for contraceptive sales to identify the impact of the pill. The third one makes use of the findings of the former two and uses sales bans to isolate the impact of the Pill on marital fertility in the 1960s. To increase readability of the causal graphs and since the strategy for the first, second and third regression is the same, I will show one causal graph for them. The graph can be applied to any of the subsequent specified models by changing the outcome variable $Y$.\
Transforming the set-up to economic language, having sales bans is seen as the *treatment* for an individual. Since sales bans are state wide laws, individuals can be grouped in living in a state with a sales ban or living in a state without sales bans. Hence, the assignment of the treatment is done by choosing one's region to live in. The problem Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) had faced, as many economic researcher do, was that it is impossible to obtain the counterfactual outcome of individuals that live in states that had sales bans in force and states that had no sales bans in force. The latter is not an issue, if the individuals are assigned randomly to the different states and, thus to the treatment. One could argue that the sales bans are random, since (most) individuals cannot affect the decision of the government whether to introduce sales bans or not. However, due to two main arguments one cannot assume that the treatment is randomly assigned to individuals. 

- Individuals could choose the state they live in according to the freedom of using contraceptive methods. Such individuals might be more likely to use contraceptive methods.
- The decision whether a state has sales bans for contraceptive methods or not could reflect the general attitude of individuals living in a state. More conservative states might tend to have sales bans, while having habitants tending to use less contraceptives.

The causal graph below shows the relationships between the treatment variables *sales bans*, the assignment variable *state*, the unobservable variable *time* (in a sense that its effect is unobservable) and the outcome variable *Y*. In addition, there are *controls* and *other unobservables*. The latter are assumed to be uncorrelated with the other regressors. The channel from time to sales bans directs through the Griswold decision, which canceled most laws regarding sales bans from 1965 onward.

Since it is not possible to randomize groups and there are backdoor paths through the unobservable *state* and *time* variables, it is not possible to only condition on the *controls*. As derived in the methodology section, the DiD regression adds time-, state-fixed effects to block the respective backdoor paths. However, there is still the path that directs from time through the Griswold decision to sales bans. This is where the interaction term jumps in to block this additional path.\
Next, I will show that the parameter $- \gamma_{int}$, derived in the methodology section, has a causal interpretation, if it is possible to assume that states with sales bans in force in the first period, would have evolved equally, as states without sales bans in force, without the sales bans. To make it possible to distinguish the actual outcome and the counterfactual outcome, I denote by $y_{it}^*(1)$ the outcome of an individual, if it has been treated and by $y_{it}^*(0)$ the outcome of an individual, if it has not been treated. Regarding the presented set-up, a researcher would be interested in the difference 

$$E(y_{it}^*(1) - y_{it}^*(0)| D_{g_T} = 1, D_{t_2} = 0, x_{it}), \tag{4.1}$$

which is to say, the difference of an individual that has received treatment and its counterfactual outcome of not receiving the treatment in the first period. The key assumption of DiD is that to assume that an individual would have evolved equally, before the treatment was taken away, as an individual that has not been treated, if it had not been treated. This assumption is called *common trend assumption* and is the key assumption for causal identification in the DiD framework. In mathematical terms, it can be expressed by

\begin{align}
E(y_{it}^*(0)| D_{g_T} = 1, D_{t_2} = 0, x_{it}) -& E(y_{it}^*(0)| D_{g_T} = 1, D_{t_2} = 1, x_{it}) = E(y_{it}^*(0)| D_{g_T} = 0, D_{t_2} = 0, x_{it}) - E(y_{it}^*(0)| D_{g_T} = 0, D_{t_2} = 1, x_{it}). \tag{4.2}
\end{align}

Note that in the DiD standard case, it is usually derived the other way around and it is assumed that the groups would have evolved equally after the treatment was introduced. Armed with the *common trend assumption* and some algebra it is shown in the following that the DiD method provides a framework to interpret $- \gamma_{int}$ as a causal effect, that measures what happens, if the treatment is canceled.

\begin{align}
& E(y_{it}^*(1) - y_{it}^*(0)| D_{g_T} = 1, D_{t_2} = 0, x_{it}) \\
=& \left(E(y_{it}^*(1)| D_{g_T} = 1, D_{t_2} = 0, x_{it}) - E(y_{it}^*(0)| D_{g_T} = 1, D_{t_2} = 0, x_{it})\right) - \left(E(y_{it}^*(0)| D_{g_T} = 1, D_{t_2} = 1, x_{it}) - E(y_{it}^*(0)| D_{g_T} = 1, D_{t_2} = 1, x_{it})\right)\\
=& \left(E(y_{it}^*(1)| D_{g_T} = 1, D_{t_2} = 0, x_{it}) - E(y_{it}^*(0)| D_{g_T} = 0, D_{t_2} = 0, x_{it})\right)- \left(E(y_{it}^*(0)| D_{g_T} = 1, D_{t_2} = 1, x_{it}) - E(y_{it}^*(0)| D_{g_T} = 0, D_{t_2} = 1, x_{it})\right) = - \gamma_{int} \tag{4.3}
\end{align}

For the second equality sign, it is made use of the *common trend assumption*. A common approach to test whether this assumption holds is to plot pre-treatment periods and see, if the different groups have evolved equally (Angrist and Pischke, [2009](https://www.researchgate.net/publication/51992844_Mostly_Harmless_Econometrics_An_Empiricist%27s_Companion)). I do so in the extensions in section 7.

## 5.2.1. Replication of 'Did Sales Bans Slow the Diffusion of Oral Contraception?'
For the first regression analysis Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) uses only the 1965 and 1970 NFS surveys. Furthermore, she only includes married white woman between 18 and 39, since only this kind of data is present in all three surveys. She then applies the introduced Difference-in-Differences Probit model to the pooled survey data. I slightly deviate from the notation in the paper to be more in line with the notation of the probit methodology section. The main equation Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) estimates is the following

$$ P(C_{ist} = 1 | \text{covariates}) = \Phi_{st}[1(Salesban_s = 1) f_t'\tau_1 + 1(Exception_s = 1) f_t'\tau_2 + 1(Advertising_s = 1) f_t'\tau_3 + g_{rt} + X'_{ist}\beta]  \tag{5.1}$$

- $C_{ist}$ is a dummy, that is one, if individual $i$ in state $s$ observed in year $t$ ever used the Pill
- 1(Salesban_s = 1) indicator function, if there were Sales bans in state $s$.
- 1(Exception_s = 1) indicator function, if state $s$ had a codified physician exception.
- 1(Advertising_s = 1) indicator function, if state $s$ had no advertising ban for contraception
- $f'_t := [1, 1(t = 1970)]$ is a vector where the first element is one, representing a constant term, and the second element is one or zero, indicating whether an observation is from the year 1970.
- $\tau_1' = [\tau_{1,1}, \tau_{1,1970}]$ are the coefficients of interest
    - $\tau_{1,1}$ is the point estimate of the $Salesban_s$ dummy 
    - $\tau_{1,1970}$ is the point estimate of the interaction of $Salesban_s$ and the indicator for observations from 1970 
- $g_{rt}$ are regional fixed effects
- $X$ includes an intercept, age, religion, education, husband's income and ideals about children. 

Using the results of the methodology section about the DID probit model, $\tau_{1,1}$ can be interpreted as the average difference in the z-score of Pill use in states with sales bans and the control group in 1965 and $\tau_{1,1970}$ captures how this changed from 1965 to 1970. In the analysis Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) estimated four different specifications. They differed by leaving out some of the variables from the main equation. Hence, the control groups and the control variables differed, which enhances, if the results are similar, the robustness of the results. 
The first model only includes the *Sales* term (dummy and interaction), the physical exception term *(PX)* and census region dummies. Hence, the control group is defined as the states without restrictive sales bans and no physician exception. In the second model $1(Advertising_s = 1) f_t'$ is added to the model, making the control group states with no physician exceptions and advertising bans only. The third and fourth model specification add different covariates specified in $X$.

The following function excludes the observations such that the data set fits the one used in the paper. Furthermore, it generates the relevant dummy variables and interaction terms for the analysis.

## 6.2 Replication of 'The Relationship between Sales Bans and Period Fertility'
---

The dependent variable in the linear panel equivalent of the previous analysis is the general fertility rate (GFR), measured as an Index with basis year 1965, for each state $s$ and year $t$. The variables of interest and the covariates are similar to the DiD probit model specification. However, the time horizon expands from only two periods to 31 periods. The main equation is

$$ GFR_{st} = 1(Salesban_s = 1) f_t'\tau_1 + 1(Exception_s = 1) f_t'\tau_2 + 1(Advertising_s = 1) f_t'\tau_3 + g_{rt} + X'_{st}\beta + \epsilon_{st},\tag{6.1}$$

$t = 1950, 1951, \dots, 1980$. 
The dummies are defined analogously to the previous sections and $g_{rt}$ is again a regional fixed effect. The vector $f'_t$ becomes $[1, 1(t=1951), 1(t=1952), \dots, 1(t=1980)]'$ and **$X$** includes time-varying covariates obtained by linear interpolation of variables between census years (proportion of state population of 15 to 44 residing on a farm, currently married, nonwhite, foreign born, in poverty and mean total income as well as mean educational attainment). The regressions are done once with and once without state fixed effects. The former are included with a dummy for each year, leaving out 1950 to circumvent perfect multicollinearity. The coefficients in $\hat{\tau_1} = [\hat{\tau}_{1,1951}, \hat{\tau}_{1,1952}, \dots, \hat{\tau}_{1,1980}]'$ capture changes in the gap in birth rates in states with sales bans against the comparison groups of the respective years. Note that in the case without state-fixed effects, there is a possible violation to the identification strategy presented in section 4, since one backdoor path is not closed.

The corresponding original results of Figure 4 in Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) were computed using time fixed effect least squares with clustered robust standard errors and analytical weights. As weights the population size was used. The clustered standard errors in Stata are computed using the Liang and Zeger estimator [Stata Support](https://www.stata.com/support/faqs/statistics/standard-errors-and-vce-cluster-option/). Contrarily, Statsmodels has the Eicker-Huber-White estimator implemented [Statsmodels documentation](https://bashtage.github.io/linearmodels/doc/panel/mathematical-formula.html). Hence, the estimators might differ in small magnitude. Both estimators are discussed, inter alia, in Abadie et al. [(2017)](https://economics.mit.edu/files/13927).\
Unfortunately, the provided function for weighted linear models in Statsmodels, does not provide all features I need regarding clustered standard errors. Hence, I reshaped the given model by multiplying both sides of the given equation with the square root of the population number in the respective state and time. The implementation was straightforward using a diagonal matrix including the square roots of the population numbers on the diagonal. It can be seen in detail in the [axuliary](https://github.com/HumanCapitalAnalysis/microeconometrics-course-project-manuhuth/tree/master/auxiliary) *reg_fig4*. Since I apply the same method in the extension section and I want to keep the replication section clean, I have given a short explanaition in subsection 7.2. Subsequent, I ran, as Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98), a clustered least squares regression on the transformed data. Since the paper only reports the plots, I also ran the Stata script provided by Bailey to ensure that my results fit and that I did not do any computational mistake. For the models without state fixed effects, the point estimates exactly match the results from Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98). The standard errors are very similar and only deviate at the third digit, which might be due to the different implementations of clustered standard errors. For the model with state fixed effects, the Stata code provided by the author yields similar results as I obtained for $\hat{\tau_1}$. However, in Stata four dummies are omitted due to computational multicollinearity. This was not the case for the results I obtained with Python. Even though, the matrix which had to be inverted to obtain the least squares estimate, was close to be singular. Hence, I suspect that the default values when a matrix is classified as being computationally singular, are different between both programs. 

Since there are 30 interaction term coefficients which are of major interest, I follow Bailey (2010) and plot the results for this coefficients, along with their $95\%$ confidence intervals, rather than reporting them in a table. Since there are many lines, I have decided to draw any specification in a different color. The coefficients are the solid lines and the lower and upper bounds of the $95\%$ confidence intervals are the dashed lines. The vertical blue line indicates the year 1957, the year of the introduction of the Envoid. The vertical green line indicates the year of the Griswold decision ind 1965. 

The first graph includes three specifications without state fixed effects:

First, note that a parameter is significant at the $95\%$ level, if the horizontal line at zero is not between the upper and lower bound of the respective points of the confidence interval. The results obtained throughout the different specifications differ only little in magnitude and have the same shape. This is interesting since it suggests that there is no backdoor path that needs to be closed by including state-fixed effects. In the years from 1950 to 1957 the point estimates for the average within-region difference between states with sales bans and the comparison groups remain relatively stable around zero indicating a non-statistically significant difference. ## 7.1 Extension | Checking the Common Trend Assumption
---

To explore the validity of using the DiD method, I check the common trend assumption if it is possible. Since the surveys, used to build the data, only asked for the first time of pill use but not for thefirst time of pill use of other contraceptive methods, it is unfortunately only possible to check this assumption for Table 2. 

### 7.1.1 Extension | Common Trend Assumption for Table 2 - (Changes in Birth Control Pill Use in States with Sales Bans)

Since the latent variable of the DiD probit model $y_{it}^*$ is unobservable in the real world, checking the common trend assumption cannot easily be done by comparing the (conditional) levels of $y_{it}^*$. Recall that the indicator variable $y_{it}$ takes the value one if $y_{it}^* > 0$ and otherwise it takes the value zero. My strategy is now to approximate the trends of $y_{it}^*$ with the trends of $y_{it}$. However, since the number of women differ one cannot only plot the cumulative number of women from states with sales bans in one region that have ever used the pill over time against the cumulative number of women that have never used the pill from the same sample. My solution is to divide the cumulative number in one region by the total number of individuals residing in the corresponding region. Using the proportion is in line with the Probit approach from Table 2, which models probabilities. Moreover, what a researcher is interested in, is that the conditional common trend is equal, see equation *(4.2)*. To approximate at least conditioning on the region, I divide the sample in one graph for each state and also show the pooled graph. I heavily emphasize that my approach makes use of this simplifications. However, it should give a first glimpse, whether to trust the common trend assumption or not in the regression of Table 2. 

As derived in sections 3 and 4, with a holding common trend assumption and a significant treatment effect, the trend should differ for the pre 1965 period and be equal for the post 1965 period. As stated in section 5, the south only has a few observations for which sales bans are allowed. Thus, the results for the south might no be very reliable. All regions started with around zero percent of pill users in 1960. For the Northeast the levels of the proportion was equally until the Griswold decision in 1965, then the states with former sales bans took off, whereas states without former sales bans had a much fewer increase. The Midwest and the West have a similar shape for states with sales bans and for states without sales bans in basically every year. Even though this strengthens the evidence that the common trend assumption holds, the similarities would contradict that the sales bans actually had an impact of the pill use. The south has a rapid drop in the increase in pill users around the Griswold decision. This is remarkable, since this would not match the expectation of easier access to oral contraception. However, this could be driven by the small number of observations. \
The pooled sample shows that the overall proportion of pill user increases similar until the Griswold decision and than experiences a slightly higher increase for the states with former sales bans.

My findings do not enhance the idea that the common trend assumption holds and there is a treatment effect. Either the post periods have different shapes, or the shape of the evolvement is similar for every period. However, it should be remembered that it is not possible to graphically observe the *conditional* common trend assumption and that I am not able to observe the true latent process $y_{it}^*$.

## 7.2 Extension | Using the Multi Periodic DiD Approach to Estimate the Effect of the Griswold Laws 
---

In this subsection I try to quantify the influence of sales bans on the general fertility rate. For this purpose I have modified the set up from subsection 6.2. My regression approach builds on the *classical* multi periodic DiD discussed, inter alia, in Angrist and Pischke [(2009)](https://www.researchgate.net/publication/51992844_Mostly_Harmless_Econometrics_An_Empiricist%27s_Companion). In Bailey 's analysis, the variable *sales* is used as an indicator whether a state had sales bans or not. Hence, the variable does not change over time but determines the group of an individual or state. This is reasonable, since Bailey examines the differences in those states using the Griswold decision in 1965 as a benchmark point. My approach is to define a new variable that equals one, if a state $s$ had sales bans in period $t$ and zero otherwise. This strategy allows to quantify the effect of sales bans. I will outline how exactly I created the variable above the subsequent respective programm code.
I incorporated state fixed effects, time fixed effects, age group specific effects, the same covariates as in subsection 6.2 and a DiD term. Hence, the identification strategy is the same as outlined in section 4. The DiD term $DiD_{st}$ is one, if a state $s$ had sales bans in period $t$ and zero otherwise. The underlying model is therefore

\begin{align}
GFR_{st} = \gamma_s + \tau_t + \delta_{did}\cdot DiD_{st} + \alpha_{st} + X'_{st}\beta + \varepsilon_{st}. 
\tag{7.1}
\end{align}

The estimated treatment effect can be derived as seen in sections 3 and 4. Note that the set-up is again the reverse DiD set up with the estimated effect of canceling the treatment of $-\hat{\delta}_{did}$. To implement the fixed effects, I used one-hot-encoding and always left out the first column to circumvent perfect multicollinearity with the constant. 

```python
#gen state dummies
state_dummies = [] #set up empty list for state dummy names. used later in the regression
states = df_pfer['statename'].unique() #get all names of states
for s in states: #loop to create the dummies
    name_help = 'd{}'.format(s) #define name of dummies as d*Statename* and store them in a helper object
    df_pfer[name_help] = 0 #create dummy with only zero entries
    df_pfer.loc[ df_pfer['statename'] == s, name_help] = np.repeat(1, len(df_pfer.loc[ df_pfer['statename'] == s, name_help])) #replace zeros with ones if the observaton is from state s
    state_dummies.append(name_help) #update state dummy list

state_dummies.remove(state_dummies[0])  #remove first state from the name list to circumvent dummy variable trap in regression
    
#gen year dummies 
year_dummies = [] #set up empty list for yeardummy names. used later in the regression
years = df_pfer['year'].unique() #get all years
for t in years: #loop to create the dummies
    name_help = 'd{}'.format(t) #define name of dummies as d*Year* and store them in a helper object
    df_pfer[name_help] = 0 #create dummy with only zero entries
    df_pfer.loc[ df_pfer['year'] == t, name_help] = np.repeat(1, len(df_pfer.loc[ df_pfer['year'] == t, name_help]))  #replace zeros with ones if the observaton is from year t
    year_dummies.append(name_help)  #update year dummy list

year_dummies.remove(year_dummies[0])  #remove first year from the name list to circumvent dummy variable trap in regression
   
df_ext_reg = df_pfer[ df_pfer['_agegroup'] !=0 ]    
    
#gen age dummies
age_dummies = [] #set up empty list for agegroupdummy names. used later in the regression
ages = df_ext_reg['_agegroup'].unique() #get all agegroups
for a in ages: #loop to create the dummies
    name_help = 'd{}'.format(a) #define name of dummies as d*Agegroup* and store them in a helper object
    df_ext_reg[name_help] = 0 #create dummy with only zero entries
    df_ext_reg.loc[ df_ext_reg['_agegroup'] == a, name_help] = np.repeat(1, len(df_ext_reg.loc[ df_ext_reg['_agegroup'] == a, name_help]))  #replace zeros with ones if the observaton is from the agegroup a
    age_dummies.append(name_help) #update agegroup dummy list

age_dummies.remove(age_dummies[0])  #remove first agegroup from the name list to circumvent dummy variable trap in regression
```

To create the treatment variable $DiD_{st}$, I first set it equal to the *sales* variable, which is equal to one if the states had sales bans at the beginning of the 1960. As stated above, it was equal to one in any period in the previous analysis, if a state had a sales ban. Since Arizona (1962), Colorado (1961), Illinois (1961), Indiana (1963), Nevada (1963), New Jersey (1963) canceled their laws against the sales for oral contraception already before 1965, I have set the $DiD_{st}$ variable to zero in the respective state, when $t$ was larger or equal to the respective year. I did the same for Connecticut, the state of the Griswold decision, for years after 1965 onward. Unfortunately, I have no data about when the other states canceled these bans. Hence, I assume for my analysis that this happened in 1966, since Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) argues that even thought states still had laws against oral contraception, they were not active anymore. Furthermore, I have tried the years 1966, 1967, 1968, 1969 and 1970 years and all of them yielded qualitatively similar results.

```python
#set up DiD variable
df_ext_reg['did'] = df_ext_reg['sales'] #first set up did variable to be equal to sales available
df_ext_reg.loc[df_ext_reg['year'] >= 1966, 'did'] = np.repeat(0, len(df_ext_reg.loc[df_ext_reg['year'] >= 1966, 'did'] )) #set up cut off point after Griwsold decision
states_change = ['Arizona', 'Colorado', 'Connecticut', 'Illinois', 'Indiana', 'Nevada', 'New Jersey'] #define states that experienced change before 1966
years_change = [1962, 1961, 1965, 1961, 1963, 1963, 1963] #define year of change of corresponding state

for k in range(0, len(states_change)): #run loop to adjust did according to years of given years
    df_ext_reg.loc[ (df_ext_reg['statename'] == states_change[k]) & (df_ext_reg['year'] >= years_change[k]), 'did' ] = np.repeat(0, len( df_ext_reg['did'][ (df_ext_reg['statename'] == states_change[k]) & (df_ext_reg['year'] >= years_change[k]) ]))
```

I run my model as a weighted least squares regression. To show how I implemented the latter, I introduce matrix notation of equation (7.1). Let $n$ be the number of observations and $p$ be the number of regressors including the constant. $G \in \mathbb{M}_{n \times 1}$ is the vector containing all observations for the general fertility rate, $Z \in \mathbb{M}_{n \times p}$ is a matrix containing *all* variables, $\beta \in \mathbb{M}_{p \times 1}$ is the vector of parameters of $Z$, $W \in \mathbb{M}_{n \times n}$ is a diagonal matrix with the square root of the population size on the diagonals and $\varepsilon \in \mathbb{M}_{n \times 1}$ is the vector of the errors. The equation to estimate is therefore

\begin{align}
    \underbrace{WG}_{G^*} = \underbrace{WZ}_{Z^*}\beta + \underbrace{W \varepsilon}_{\varepsilon^*} \leftrightarrow G^* = Z^*\beta + \varepsilon^*
    \tag{7.2}
\end{align}

Hence, the problem can be solved using least squares for the transformed variables
\begin{align}
\hat{\beta} = \arg \min_{b \in \mathbb{M}_{p \times 1}} ||G^* - Z^*b||^2_2
\tag{7.3}
\end{align}

```python
#define independent and dependent variables
Y = df_ext_reg['_fert'] #define dependent variable

Z = df_ext_reg[ ['did'] + year_dummies + state_dummies + ['_farm', '_curmar', '_Nonwhite', '_fb', '_higrade', '_inctot', '_poverty'] + age_dummies ] #merge covariate names
Z.loc[:,'const'] = np.repeat(1, len(Y)) #create constant

#measure by population weight
W = np.sqrt(np.diag(df_ext_reg['pop']))  #define weight matrix by population size
WY = np.dot(W,Y) #create weighted dependent variable
WZ = np.dot(W,Z) #create weighted covariates
WZ = pd.DataFrame(WZ, columns = Z.columns) #name columns of covariates

#run regressions
model1 = sm.OLS(WY,WZ) #define model
reg1 = model1.fit(cov_type='cluster', cov_kwds={'groups': df_ext_reg['statefip']}) #run clustered model

#define data frame to show results
df_display = pd.DataFrame(columns = ['', 'weighted'])
df_display['']  = ['DiD', ' ', 'p-value']

df_display.loc[0, k] = np.round(reg1.params['did'], 4)
df_display.loc[1, k] = '({})'.format(np.round(np.sqrt(reg1.cov_params_default[0,0]), 4))
df_display.loc[2, k] = np.round(reg1.pvalues['did'], 4)      
```

The coefficient estimate $\hat{\gamma}$ is positive but not significant at the $5\%$ level, since its p-value is $0.0534$. With what is derived for the reverse DiD, the effect of canceling the treatment is estimated to be $-3.2046$. Hence, states with sales bans are expected to have a higher GFR index of $3.2046$. However, since it cannot clearly be determined whether the parameter is different from zero, it is hard to argue from this regression, that the sales bans really had an influence on period fertility.

## 7.3 Extension | Using Different Starting Values to Check the Robustness of the Results
---

As outlined in the methodology section, the results for the DiD Probit models in section 5 are derived using a numerical optimizer, since there is no closed form solution. For this purposes it is necessary to choose an appropriate numerical method and starting values to find a global optimum. The question that comes along with that is how to choose the *appropriate* technique and starting values. Most software packages, including Stata and Python, have default values. The default values for the Probit model in Python are zero for all parameters [(Statsmodels Description)](https://www.statsmodels.org/stable/generated/statsmodels.discrete.discrete_model.Probit.fit.html#statsmodels.discrete.discrete_model.Probit.fit). Depending on the shape of the maximum log likelihood (MLL) function, it cannot be guaranteed that one has reached a global maximum using these values. Especially, when the MLL is very flat, it can happen that the algorithm is very sensitive to the chosen starting values. The problem for the researcher is that in most economic applications the structure of the MLL is, due to the high amount of parameters, hard to track and determine. One solution is to use different starting values and optimizing techniques to determine how stable the results are. If the results do not differ qualitatively, it gives evidence that the results are stable and a proper global maximum was found. \
Note that the starting values directly influence the optimization problem of equation *(3.3)* but the reported parameter estimates are the AME, which are dependent on the parameter estimates of *(3.3)* as outlined in equation *(3.4)*.

### 7.3.1 Extension | Theory for Random Draws of the Starting Values

My strategy is to apply random starting values for all parameters and for every model to derive the distribution of the parameters of the *sales* variable and its interaction terms depending on the starting values, conditioning on the data. The parameters are drawn from a uniform distribution. Since I want to circumvent numerical instability resulting in NaNs during the optimization process, the choice of the upper and lower bounds of the random draw is important. My strategy is to have a look at the different maximal and minimal values of the parameters, **not** the marginal parameter, from the regression with starting values of zero for all parameters.
I define the maximum and minimum value of the parameter of regression 5.2.1/5.2.2 by

\begin{align}
    m_{upp}= \max \{ \hat{\beta} \mid \hat{\beta} \text{ is an estimated parameter from the respective subsection} \} \\
    m_{low}= \min \{ \hat{\beta} \mid \hat{\beta} \text{ is an estimated parameter from the respective subsection} \}
    \tag{7.4}
\end{align}

and set the upper and lower bounds to

\begin{align}
    b_{upp} =& \max\{m_{upp} + \lambda \cdot |m_{upp}|, 1\} \\
    b_{low} =& \min\{m_{low} - \lambda \cdot |m_{low}|, -1 \}. 
    \tag{7.5}
\end{align}

To specify a minimum of variability, I set the bounds such that the random coefficients can at least vary between $-1$ and $1$. To restrict the size of the coefficients, I decided to allow for deviations of two times the absolute value. Thus, I chose to circumvent very high starting values, resulting in NaNs in the log-likelihood function, which would make the number of needed iterations increasing. Hence, there is a trade-off between more variability and computational costs by choosing the right bounds. Since this issue is not of greater concern for my validity check, I do not conduct further analysis on this, even though this would be needed for further analysis.

### 7.3.2 Extension | Using Different Starting Values to Check the Robustness of the Results of Subsection 5.2.1

First I define the models which are to be examined and compute the upper and lower bounds according to formulas 7.4 and 7.5.

```python
#define the models to be examined
models1 = [sec5_2_1_model1, sec5_2_1_model2, sec5_2_1_model3, sec5_2_1_model4]

lamb = 2 #set the lambda multiplier to adjust the values of the upper and lower bound function

min_lst = [] #set up empty list for the respective minimum values of all regressions
max_lst = [] #set up empty list for the respective mmaximum values of all regressions
for m in models1:
    fit = m.fit(cov_type='cluster', cov_kwds={'groups': df_reg1['_region']}, disp = False) #get the nonmarginal coefficients of model m
    min_lst.append(np.min(fit.params)) #store minimum value
    max_lst.append(np.max(fit.params)) #store maximum value

m_low = np.min(min_lst) #compute minimum of all regressions according to formula 7.4
m_upp = np.max(max_lst) #compute maximum of all regressions according to formula 7.4
b_low = min([m_low-lamb*abs(m_low), -1]) #compute the lower bound according to formula 7.5
b_upp = max([m_upp+lamb*abs(m_upp), 1])  #compute the upper bound according to formula 7.5
print('The values of estimated parameter for all models in section 5.2.1 range from {} to {}'.format(round(m_low,4), round(m_upp,4)) )
print('The upper and lower bounds are {} and {}'.format(round(b_low,4), round(b_upp,4)) )
```

# 8. Conclusion and Critical Assessment
---

In general, my findings support the results from Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98). The available detailed documentation of Stata do-files allowed me to precisely replicate the paper and to facilitate intuition about potential differences of the results. Most parameter estimates of the AME in the probit models and marginal effects in the linear panel model are the same or only differ small in magnitude, which do not change the results qualitatively and could be due to computational issues. However, for Table 2 the clustered standard errors differ remarkably for the second model specification. As discussed, I suspect changes in Stata, maybe a transfer error or different implementations of the marginal standard errors.\
Additionally to the replication, I examine the identification strategy using a causal graph and discuss the properties of the DiD (probit) model. Especially, in context of the theoretical framework of Michael and Willis [(1976)](https://www.nber.org/chapters/c3960.pdf). The work in Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) enhances this model by giving empirical evidence that the introduction of the pill reduced the number of born children, controlling for a constant demand for children and assuming strict enforcement of the Comstock Laws. The findings in Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) support the theoretical model by finding that the sales bans had an influence on the total cost function of an individual. 

In my extension section, I tried to validate the robustness of the results by checking the common trend assumptions for sections five and six, establishing a model to quantify the effect of the sales bans and using different starting values for the probit specifications. The common trend assumption is not validated using my results of the Probit model. However, since it is impossible to check the conditional trends and to observe the latent probit process, I want to remark that my findings do not give full evidence that the assumption is violated. The common trend assumption of the linear panel model, even though only conditioned on regional level, seems to be met by my results. The new established model and the simulation for random starting values enhance the results of Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98). However, the study has probably to deal with some bias, since its data comes from surveys and individuals had to indicate whether they have used the pill, even though it was illegal at the time of us. This bias is hardly possible to quantify. However, I would suspect that less women have indicated that they have used the pill as they actually did in states with sales bans. This could be neglectable if the number is small. If not, the direction of the bias is not clear, since it is not known which individuals have indicated wrong answers.\
The validity check using different starting values and optimizers showed that the results only converged reliable for the analysis in subsection 5.2.1 and one optimizer. Hence, my results give evidence that the estimates in the paper might be sensitive to the chosen optimizer and starting values. Further research on this would be desirable. 

Overall, Bailey [(2010)](https://pubs.aeaweb.org/doi/pdfplus/10.1257/aer.100.1.98) is able to find quite robust evidence that the introduction of the pill reduced marital childbearing in the US in the 1960s and the subsequent years. She contributes to the literature by a new empirical strategy evaluating the role of the pill without discounting other forces of the 1960s like an increase in working women, less gender discrimination or changing norms.
